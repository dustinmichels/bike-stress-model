```python
import geopandas as gpd
```

```python
census_blocks = gpd.read_file("../data/raw/Census_2020_Blocks.zip")

# read the Somerville boundary GeoJSON
somerville_boundary = gpd.read_file("../data/out/somerville_boundary.geojson")

# make sure both GeoDataFrames have the same CRS
somerville_boundary = somerville_boundary.to_crs(census_blocks.crs)

# clip blockgroups to the Somerville boundary
somerville_census_blocks = gpd.clip(census_blocks, somerville_boundary)
```

```python
somerville_census_blocks['centroid'] = somerville_census_blocks['geometry'].centroid
```
